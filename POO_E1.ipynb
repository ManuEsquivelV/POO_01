```python
#1 Crea el arreglo
import numpy as np
x = np.arange(10)
```

```python
#2 Realiza una segmentación para mostrar los valores de 4 a 7
segmentacion = x[4:8]
print("Valores de 4 a 7:", segmentacion)
```

```python
#3  Usa el atributo np.split(x,[3,6]). Para x1, x2,x2 = np.split(x,[3,6]). ¿Qué valores te muestra?
x1, x2, x3 = np.split(x, [3, 6])
print("x1:", x1)
print("x2:", x2)
print("x3:", x3)

```

3.-
* "x1" estará compuesto por los elementos desde el principio hasta el tercer índice.
* "x2" constará de los elementos desde el tercer índice hasta el sexto índice sin tomarlo en cuenta.
* "x3" estará formado por los elementos desde el sexto índice hasta el final.

7.- Explica qué operación realiza el atributo .drop.duplicate()

Este se usa para buscar un subconjunto de columnas o filas dentro de una tabla de reslutados que estén duplicados para poder simplificarlas.

```python
#8 Explica, cuidadosamente, el data frame:

import pandas as pd
import numpy as np

data = pd.DataFrame(np.arange(12).reshape((3, 4)),
                    index=['CDMX', 'Guanajuato', 'Puebla'],
                    columns=['uno', 'dos', 'tres', 'cuatro'])

print(data)
```

8.- Crea un DataFrame:

*   Reshape = crea una Matriz de datos.
*   Index = crea los indices de filas.
*   Columnas = crea los nombres de columnas.

```python
#9  Del dataframe anterior, renombre las etiquetas que provee Python sobre los renglones.
data = data.rename(index={'CDMX': 'Ciudad de Mexico', 'Guanajuato': 'Gto', 'Puebla': 'Pue'})

print(data)
```

```python
#10 Explica el siguiente código:

import pandas as pd
import numpy as np

df = pd.DataFrame({'key1': ['a', 'a', 'b', 'b', 'a'],
                   'key2': ['one', 'two', 'one', 'two', 'one'],
                   'data1': np.random.randn(5),
                   'data2': np.random.randn(5)})
print(df)
```

```python
# 10A Quita la columna de "key 2".

df = df.drop('key2', axis=1)
df
```

```python
#10b Cambia el índice a tu conveniencia.

df = df.set_index('key1')
df
```

```python
#10C Realiza un análisis de estadística descriptiva (No es necesario explicar cada resultado).

descripcion = df.describe()
descripcion
```

```python
#10D Calcula la media de cada columna sobrante.

medias = df.mean()
medias
```
