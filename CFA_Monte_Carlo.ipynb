```python
import yfinance as yf
import pandas as pd
import numpy as np

# Define el ticker de la acción de la BMV
ticker = 'BOLSAA.MX'  # Reemplaza 'TICKER' con el símbolo de la acción que te interese

# Descargar datos históricos desde 2018 hasta la fecha actual
historical_data = yf.download(ticker, start='2018-01-01', end='2023-10-30')

# Extraer la columna 'Adj Close' que contiene los precios de cierre ajustados
historical_closing_prices = historical_data['Adj Close']

# Calcular los rendimientos diarios
daily_returns = historical_closing_prices.pct_change().dropna()

# Calcular el VaR histórico al 95% de confianza
confidence_level = 0.05
var_95 = daily_returns.quantile(confidence_level)

print("VaR al 95% de confianza:")
print(var_95)

# Realizar un análisis de sensibilidad
# Puedes hacer cambios en el precio de la acción hipotéticamente y calcular el impacto en el VaR
# Por ejemplo:
nuevo_precio = historical_closing_prices.iloc[-1] * 1.05  # Aumento del 10% en el precio
nuevos_returns = (historical_closing_prices / historical_closing_prices.iloc[-1]) * nuevo_precio
new_var_95 = nuevos_returns.pct_change().quantile(confidence_level)
sensitivity = new_var_95 - var_95
print(nuevo_precio)
print(nuevos_returns)
print(sensitivity)
```
